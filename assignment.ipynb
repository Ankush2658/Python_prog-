```python
#arrenge string in even odd index
def evenoddstring(string):
    even=[]
    odd=[]
    for i in range(0,len(string)):
        if i%2==0:
            (even.append(string[i]))
        else:
            (odd.append(string[i]))
    print('even index : ',','.join(even))
    print('odd index  :',','.join(odd))

evenoddstring(input('enter string : '))
```
